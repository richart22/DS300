# ***Products***

## ***Đọc dữ liệu và gộp dữ liệu***

```python
import pandas as pd
import numpy as np

df_innerwear = pd.read_csv('uniqlo_women_innerwear.csv', sep=';')
df_innerwear.head()
```

```python
import pandas as pd

# Lấy danh sách các file bắt đầu bằng uniqlo_women_ và kết thúc .csv
files = ["uniqlo_women_accessories.csv", 
         "uniqlo_women_bottoms.csv", 
         "uniqlo_women_outerwear.csv",
         "uniqlo_women_loungewear-and-homewear.csv",
         "uniqlo_women_shirts-and-blouses.csv",
         "uniqlo_women_skirts-and-dresses.csv",
         "uniqlo_women_sweaters-and-knitwear.csv",
         "uniqlo_women_t-shirts-sweat-and-fleece.csv",
         "uniqlo_men_accessories.csv", 
         "uniqlo_men_bottoms.csv", 
         "uniqlo_men_innerwear.csv", 
         "uniqlo_men_outerwear.csv",
         "uniqlo_men_loungewear-and-homewear.csv",
         "uniqlo_men_linen.csv",
         "uniqlo_men_special-collaboration.csv",
         "uniqlo_men_uv-protection.csv",
         "uniqlo_men_shirts-and-polo-shirts.csv",
         "uniqlo_men_sweaters-and-knitwear.csv",
         "uniqlo_men_t-shirts-sweat-and-fleece.csv"]

# Đọc và gộp các file
df_list = [pd.read_csv(file) for file in files]
df_merged = pd.concat(df_list, ignore_index=True)
df_merged = pd.concat([df_merged, df_innerwear], ignore_index=True)

df_merged.head()
```

```python
df_merged.isna().sum()
```

## ***Xử lý thuộc tính Price***

```python
df_merged['Price'] = df_merged['Price'].astype(str).str.replace('.', '', regex=False).str.strip()
df_merged['Price'] = pd.to_numeric(df_merged['Price'], errors='coerce')
df_merged
```

```python
df_merged.info()
```

```python
df_merged.isna().sum()
```

```python
def adjust_price(price):
    if 1000 <= price < 10000:
        return price * 100
    return price

df_merged['Price'] = df_merged['Price'].apply(adjust_price)
df_merged.isna().sum()
```

## ***Loại bỏ giá trị trùng lặp***

```python
duplicates_mask = df_merged.duplicated(subset=['Product Name'], keep='first')
removed_product_duplicate = df_merged.loc[duplicates_mask, 'Mã sản phẩm'].unique().tolist()
df_merged = df_merged.drop_duplicates(subset=['Product Name'], keep='first')
df_merged
```

```python
df_merged = df_merged.reset_index(drop = True)
df = df_merged.copy()
```

## ***Xử lý thuộc tính Description***

```python
df['Description'] = df['Description'].replace('Màu sắc ', '', regex=True)
df['Description'] = df['Description'].replace('Colour ', '', regex=True)
df
```

## ***Xử lý thuộc tính Size***

```python
df['Size'] = df['Size'].replace('One', 'One Size', regex=True)
df
```

## ***Xử lý thuộc tính rate***

```python
df['Rate'] = pd.to_numeric(df['Rate'], errors='coerce')
condition = (df['Rate'] > 5) | (df['Rate'].isna())
dropped_product_rate = df.loc[condition, 'Mã sản phẩm'].unique().tolist()
df = df[~condition].reset_index(drop=True)
df
```

## ***Lấy ra những ID nào đã bị xóa đi để loại bỏ trong phần reviewer***

```python
print(f"Số lượng dòng đã xóa không có đánh giá: {len(dropped_product_rate)}")
print("Danh sách Mã sản phẩm bị xóa do không có đánh giá:", dropped_product_rate)
```

```python
print(f"Số lượng dòng đã xóa bị trùng lặp: {len(removed_product_duplicate)}")
print("Danh sách Mã sản phẩm bị trùng lặp:", removed_product_duplicate)
```

```python
removed_product_id = dropped_product_rate + removed_product_duplicate
print(f"Số lượng dòng đã xóa: {len(removed_product_id)}")
print("Danh sách Mã sản phẩm:", removed_product_id)
```

```python
df
```

## ***Lưu file product***

```python
df.to_csv("uniqlo_product.csv", index=False, encoding='utf-8-sig')
```

### ***Bỏ đi những cái tên - và xem xét nếu như số lượng Reviewer trong từng Mã sản phẩm như nào rồi xử lý nó***

```python
products_info = df_reviewer.loc[df_reviewer['Reviewer'].eq('-'), ['Product Name', 'Mã sản phẩm']]
tong = 0

lost_review_product_ids = []
for index, row in products_info.iterrows():
    prod_name = row['Product Name']
    prod_id = row['Mã sản phẩm']
    
    mask = df_reviewer['Mã sản phẩm'].eq(prod_id) & df_reviewer['Reviewer'].ne('-')
    count = mask.sum()
    
    if count == 0:
        print(f"Mã: {prod_id} | Tên: {prod_name}")
        lost_review_product_ids.append(prod_id)
    tong += count
    

print(f"---")
print(f"Tong: {tong}")
```

```python
df_reviewer = df_reviewer[df_reviewer['Reviewer'] != "-"]
df_reviewer = df_reviewer.reset_index(drop=True)
```
