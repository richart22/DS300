```python
# Identify Product Duplicates (based on ID)
prod_keys = ['Mã sản phẩm']
product_dupes = df_products[df_products.duplicated(subset=prod_keys, keep=False)].sort_values(by=prod_keys)

print("--- Product Duplicates Found ---")
# Showing specific columns to highlight the subtle name differences
product_dupes[['product_url', 'Mã sản phẩm', 'Product Name', 'category']]
```

```python
# Identify Review Duplicates (based on User, Product, and Date)
rev_keys = ['Reviewer', 'Mã sản phẩm', 'Review Date']
review_dupes = df_reviews[df_reviews.duplicated(subset=rev_keys, keep=False)].sort_values(by=rev_keys)

print("\n--- Review Duplicates Found ---")
review_dupes[rev_keys + ['Rating']]
```

```python
# 1. Drop duplicates for Product Data based on ID
# We keep 'first' to maintain one version of the product name/metadata
product_df_cleaned = df_products.drop_duplicates(subset=['Mã sản phẩm'], keep='first')

# 2. Drop duplicates for Review Data
# We remove exact duplicates of (User, Product, Date)
review_df_cleaned = df_reviews.drop_duplicates(subset=['Reviewer', 'Mã sản phẩm', 'Review Date'], keep='first')

# 3. Save the "Gold Standard" datasets for ML/GNN training
product_df_cleaned.to_csv('uniqlo_product_cleaned.csv', index=False)
review_df_cleaned.to_csv('uniqlo_review_cleaned.csv', index=False)
```

```python
import os
import pandas as pd
import requests
from tqdm import tqdm

# Create the target directory
folder_name = 'product_img'
if not os.path.exists(folder_name):
    os.makedirs(folder_name)
    print(f"Created folder: {folder_name}")

# Download Logic
def download_images(dataframe, folder):
    local_paths = []

    print("Starting download process...")
    for index, row in tqdm(dataframe.iterrows(), total=len(dataframe)):
        url = row['img_url']
        product_id = str(row['Mã sản phẩm'])

        # Define filename and path (using .jpg as seen in URLs)
        filename = f"{product_id}.jpg"
        save_path = os.path.join(folder, filename)

        try:
            # Download image
            response = requests.get(url, stream=True, timeout=10)
            if response.status_code == 200:
                with open(save_path, 'wb') as f:
                    for chunk in response.iter_content(1024):
                        f.write(chunk)
                local_paths.append(filename)
            else:
                local_paths.append("download_failed")
        except Exception as e:
            local_paths.append("error_occurred")

    return local_paths

# Execute and update DataFrame
product_df['product_img'] = download_images(product_df, folder_name)

# 5. Save the updated CSV for your DL models
product_df.to_csv('uniqlo_product_with_images.csv', index=False)
print("Finished! New CSV saved as 'uniqlo_product_with_images.csv'")
```
